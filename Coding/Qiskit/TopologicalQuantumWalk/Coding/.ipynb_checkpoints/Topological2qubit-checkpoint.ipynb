```python
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline
from qiskit import *
from qiskit.visualization import plot_histogram
%run qft2.ipynb #for importing ccx function
```

```python
n = 2
coin = 1
classic = 3
qreg = QuantumRegister(n)
qcoin = QuantumRegister(coin)
qclass = ClassicalRegister(classic)
```

```python
qw = QuantumCircuit(qcoin,qreg,qclass)

qw.x(qreg[1])
qw.u3(0.4,0,0,qreg[0])
qft_inv(qw,qreg,swap=True)

qw.cx(qreg[0],qcoin[0])
qw.u1(pi/4,qcoin[0])
qw.cx(qreg[0],qcoin[0])
qw.cx(qreg[1],qcoin[0])
qw.u1(pi/2,qcoin[0])
qw.cx(qreg[1],qcoin[0])
qw.barrier()
qw.u1(-3*np.pi/4,qcoin[0])
qw.u1(pi/4,qreg[0])
qw.u1(pi/2,qreg[1])

qft1(qw,qreg,swap=True)

qw.cx(qreg[0],qcoin[0])
qw.u3(pi+0.4,0,0,qcoin[0])
qw.cx(qreg[0],qcoin[0])
qw.barrier()
                      
qft_inv(qw,qreg,swap=True)

qw.cx(qreg[0],qcoin[0])
qw.u1(pi/4,qcoin[0])
qw.cx(qreg[0],qcoin[0])
qw.cx(qreg[1],qcoin[0])
qw.u1(pi/2,qcoin[0])
qw.cx(qreg[1],qcoin[0])
qw.barrier()
qw.u1(-3*pi/4,qcoin[0])
qw.u1(-pi/4,qreg[0])
qw.u1(-pi/2,qreg[1])

qft1(qw,qreg,swap=True)


qw.measure(qcoin[0],0)
qw.measure(qreg[0],1)
qw.measure(qreg[1],2)


qw.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 2048
results = execute(qw, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```
