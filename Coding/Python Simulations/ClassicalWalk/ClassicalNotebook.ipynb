```python
from random import randint
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
plt.rcParams['figure.figsize'] = 11, 8
mpl.rcParams.update({'font.size': 15})
```

```python
def classicalWalk(nExp,steps):
    positions = []
    for x in range(nExp):
        k = 0
        for y in range(steps):
            a = randint(0,1)
            k+=(-1)**a
        positions.append(k)
    return positions

def multipleWalk(nExp,multSteps):
    xvecList = []
    for step in multSteps:
        xVec = classicalWalk(nExp,step)
        xvecList.append(xVec)
    return xvecList

def multProbDict(multWalkList,nExp):
    multProbs = []
    for walkList in multWalkList:
        probDict = {i:walkList.count(i)/ne for i in set(walkList)}
        lists = sorted(probDict.items())
        multProbs.append(lists)
    return multProbs

def plotMultiple(stepList,multiWalks,configVec):
    for step,walk,config in zip(stepList,multiWalks,configVec):
        x,y = zip(*walk)
        plt.plot(x,y,color=config[0],linestyle=config[1],label="Steps=%s"%step)
        plt.legend()
        plt.xlabel("Graph Node")
        plt.ylabel("Probability")
    return -1
```

```python
def standardDev(t):
    devList = []
    for i in range(t):
        x = i
        y = np.sqrt(i)
        devList.append((x,y))
    return devList

def plotStandardDev(devList):
    x,y = zip(*devList)
    plt.plot(x,y)
    plt.xlabel("Time")
    plt.ylabel("Standard Deviation")
    
t=450
stdDevList = standardDev(t)
plotStandardDev(stdDevList)
```
